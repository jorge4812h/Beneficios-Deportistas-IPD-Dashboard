```python
import pandas as pd
import numpy as np
import matplotlib as plt
import mysql.connector
```

# Extracción de datos desde MYSQL

```python

conexion=mysql.connector.connect(user="root",password="Escorpi@456",host="localhost",port="3306",database="deportistas")

query="Select * from data_deportistasbeneficiadospad_1"

df=pd.read_sql_query(query,conexion)

conexion.close()
```

# Limpieza y Transformación de Datos

## Asignación de ID del proyecto

```python
df.set_index("ITEM", inplace=True) 
df.info()
```

## Cambio de tipo de dato en columnas 

```python
objectcolumns = df.select_dtypes("object").columns

for i in objectcolumns:
    df[i]=df[i].astype("string")

df.info()
```

## Cambio nombre de columna

```python
df.rename(columns={"NOMBRE_COMPLET":"NOMBRE_COMPLETO"},inplace=True)
```

## Conversión de valores de MES de números a nombres

```python
df["MES"].unique()
```

```python
#Para realizar la corrección debemos convertir los datos a string y extraer los 2 valores de la columna (indican el mes que corresponde)
df["MESES"]=df["MES"].astype(str).str[-2:]

meses_año={"01":"ENERO","02":"FEBRERO","03":"MARZO","04":"ABRIL","05":"MAYO","06":"JUNIO","07":"JULIO","08":"AGOSTO","09":"SETIEMBRE","10":"OCTUBRE","11":"NOVIEMBRE","12":"DICIEMBRE"}

df["MES"]=df["MESES"].map(meses_año)
df
```

## Tratamiento de valores nulos o vacíos

```python
#Segun informacion del Dataset, hay deportistas nacidos en el extranjero por lo cual no cuentan con información sobre su departamento, provincia ni distrito.
df.info()
```

```python
# Reemplazamos los valores nulos o vacios de las 3 columnas con el valor "EXTRANJERO".
df[["DEPARTAMENTO","PROVINCIA","DISTRITO"]]=df[["DEPARTAMENTO","PROVINCIA","DISTRITO"]].fillna("EXTRANJERO")
df.info()
```

## Eliminación de columnas innecesarias del DataFrame

```python
df.drop(["FECHA_PUBLICACION","FECHA_CORTE","MESES"],axis=1,inplace=True)
df["DEPARTAMENTO"].unique()
```
